```python
import pandas as pd
import numpy as np
```

```python
dfo_md = pd.read_csv("/Users/Diana/Documents/escuela/msc/data/bbmp_aggregated_profiles.csv")
bb_md = pd.read_csv("/Users/Diana/Documents/escuela/msc/data/metadata/METADATA.csv")
```

```python
#restrict to only years of interest 2015 2016
dfo_md_15 = dfo_md[dfo_md.year_time == 2015]
dfo_md_15.loc[:,'year'] = "BB15" #add this column to match with metadata for QIIME2

dfo_md_16 = dfo_md[dfo_md.year_time == 2016]
dfo_md_16.loc[:,'year'] = "BB16"
```

```python
dfo_1516 = pd.concat([dfo_md_15, dfo_md_16])
```

```python
#restrict to only pressures 1,5,10,60
dfo_pressure_1 = dfo_1516[dfo_1516.pressure == 1.0]
dfo_pressure_1.loc[:,'depth_code'] = "A"

dfo_pressure_5 = dfo_1516[dfo_1516.pressure == 5.0]
dfo_pressure_5.loc[:,'depth_code'] = "B"

dfo_pressure_10 = dfo_1516[dfo_1516.pressure == 10.0]
dfo_pressure_10.loc[:,'depth_code'] = "C"

dfo_pressure_60 = dfo_1516[dfo_1516.pressure == 60.0]
dfo_pressure_60.loc[:,'depth_code'] = "D"
```

```python
dfo_1516 = pd.concat([dfo_pressure_1,dfo_pressure_5,dfo_pressure_10,dfo_pressure_60])
```

```python
dfo_1516 = dfo_1516.sort_values(by="time_string")
dfo_1516
```

```python
dfo_1516.shape
```

```python
#add week number
week15 = np.arange(2,52)
week15 = np.repeat(week15, 4)

week16 = np.arange(1,52)
week16 = week16[week16 != 12]
week16 = np.repeat(week16, 4)

week = np.concatenate([week15, week16])
week
```

```python
dfo_1516['week'] = week
```

```python
dfo_1516
```

```python
#how to join them:

#We'll make a new list to be our new column.
new_labels = []

#we can iterate through rows with this.
for _, row in dfo_1516.iterrows():
    #we can get the column value of each row by indexing on the column name
    year = row['year']
    depth_code = row['depth_code']
    week = row['week']
    #we use string formatting here to make a new label
    new_label = "{0}.{1}{2}".format(year, week, depth_code)
    #add the new label to the list
    new_labels.append(new_label)
#this will make a new column in the df called 'NewLabel' with the values
dfo_1516['name'] = new_labels
dfo_1516
```

```python
#merge 2 metadata files
metadata_bb = pd.merge(bb_md, dfo_1516, on = 'name',how='outer')
metadata_bb
```
